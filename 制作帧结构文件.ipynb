```python
import re
import json
file_path = r"E:\workspace\AutoTestTool\compress_data\MCF200T.json"
# 读取JSON文件
with open(file_path, 'r') as file:
    data = json.load(file)
```

```python
# 做一个

for frame_type_key, frame_type_value in data.items():
    # 遍历 frame_type 层
    for region_type_key, region_type_value in frame_type_value.items():
        # 遍历 region_type 层
        for row_num_key, row_num_value in region_type_value.items():
            # 遍历 row_num 层
            for col_num_key, col_num_value in row_num_value.items():
                frame_type = int(re.search(r'\d+$', frame_type_key).group())
                region_type = int(re.search(r'\d+$', region_type_key).group())
                row_num = int(re.search(r'\d+$', row_num_key).group()) 
                col_num = int(re.search(r'\d+$', col_num_key).group()) 
                FAR = {
                    "frame_type":frame_type,
                    "region_type":region_type,
                    "row_num":row_num,
                    "col_num":col_num
                }
                data[frame_type_key][region_type_key][row_num_key][col_num_key]["FAR"] = FAR
```

```python
filename = r"E:\workspace\AutoTestTool\compress_data\MCF200T_v2.json"
with open(filename, 'w') as j:
        json.dump(data, j)
```

```python

with open(filename, "r") as f:
    data = json.load(f)

# 你可以用嵌套dict保存统计结果，也可以直接打印
frame_counts = {}

for frame_type_key, region_dict in data.items():
    frame_type = int(frame_type_key.split("_")[-1])
    if frame_type not in frame_counts:
        frame_counts[frame_type] = {}
    for region_type_key, row_dict in region_dict.items():
        region_type = int(region_type_key.split("_")[-1])
        if region_type not in frame_counts[frame_type]:
            frame_counts[frame_type][region_type] = {}
        for row_num_key, col_dict in row_dict.items():
            row_num = int(row_num_key.split("_")[-1])
            if row_num not in frame_counts[frame_type][region_type]:
                frame_counts[frame_type][region_type][row_num] = {}
            for col_num_key, content in col_dict.items():
                # 有些json可能有其它非col_num_xxx的字段，这里要判断一下
                if not col_num_key.startswith("col_num_"):
                    continue
                col_num = int(col_num_key.split("_")[-1])
                frame_count = content.get("frame_count", None)
                frame_counts[frame_type][region_type][row_num][col_num] = frame_count

# 输出每个组合的 frame_count
for frame_type in sorted(frame_counts):
    for region_type in sorted(frame_counts[frame_type]):
        for row_num in sorted(frame_counts[frame_type][region_type]):
            for col_num in sorted(frame_counts[frame_type][region_type][row_num]):
                fc = frame_counts[frame_type][region_type][row_num][col_num]
                print(f"frame_type={frame_type}, region_type={region_type}, row={row_num}, col={col_num}: frame_count={fc}")
```
